# OpenCV

**Load RGB image**

```python
from google.colab.patches import cv2_imshow # import cv2_imshow to display images in colab
import cv2

img = cv2.imread("/content/superman-8256429_1280.webp",1)
img = cv2.resize(img,(400,600))
cv2_imshow(img) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
```

**Load Grayscale Image**

```python
img = cv2.imread("/content/superman-8256429_1280.webp",0)
img = cv2.resize(img,(400,600))
cv2_imshow(img) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
```

**IMREAD_UNCHANGED : Loads image as such including alpha channel**

```python
img = cv2.imread("/content/superman-8256429_1280.webp",-1)
img = cv2.resize(img,(400,600))
cv2_imshow(img) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
```

```python
cv2.destroyAllWindows()
```

**Use of Flip**

```python
img1 = cv2.flip(img,0)#it accept 3 parameters 0,-1,1
cv2_imshow(img1)
```

```python
img1 = cv2.flip(img,1)#it accept 3 parameters 0,-1,1
cv2_imshow(img1)
```

**Saving the Image**

```python
img1 = cv2.flip(img,1)#it accept 3 parameters 0,-1,1
cv2_imshow(img1)

k = cv2.waitKey(0) & 0xFF
img = cv2.imwrite("/content/output.jpg",img1)  #it accept name of image and data
cv2.destroyAllWindows()


#Following code will be used when running on local environment to save image
# if k == ord("q"):
#     cv2.destroyAllWindows()

# elif k == ord("s"):
#     img = cv2.imwrite("/content/output.jpg",img1)  #it accept name of image and data
#     cv2.destroyAllWindows()
```
